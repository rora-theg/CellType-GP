## 载入真实单细胞数据

```python
import numpy as np
from anndata import AnnData
import pandas as pd

# 读取空间表达矩阵
X = pd.read_csv(
    "/home/vs_theg/ST_program/CellType_GP/sc2spatial/DATA/sim_ctgp/pseudo_data.csv",
    index_col=0
)
# 读取真实比例矩阵
true_p = pd.read_csv(
    "/home/vs_theg/ST_program/CellType_GP/sc2spatial/DATA/sim_ctgp/true_p.csv",
    index_col=0
)

# 构建 AnnData
adata = AnnData(X=X.T)  # 注意：AnnData 默认是 cells × genes，而 CSV 是 genes × spots
adata.obsm["true_p"] = true_p.loc[adata.obs_names]

adata

```

```python
# === 6. 保存为标准 h5ad ===
adata.write_h5ad("pseudo_data.h5ad")
print("✅ Saved to pseudo_data.h5ad")
```
